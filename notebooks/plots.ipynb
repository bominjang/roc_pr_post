### Quick plots:

```python
np.random.seed(15)

mean1 = [0, -0.25] # Negative
cov1 = [[0.1, 0.001],[0.001, 0.1]]
v1 = np.random.multivariate_normal(mean1, cov1, 9900)

mean2 = [0, 0.25] # Positive
cov2 = [[0.1, 0.001],[0.001, 0.1]]
v2 = np.random.multivariate_normal(mean2, cov2, 100)


true_label = np.concatenate((np.zeros(len(v1)), np.ones(len(v2))), axis=None)
predictions = np.concatenate((v1[:, 1], v2[:, 1]), axis=None)
```

```python
plot_roc(true_label, predictions)
```

```python
plot_pr(true_label, predictions)
```

### Edit image - plot optimized points

```python
import pickle
from matplotlib import cm


cmap = cm.get_cmap('cool')

with open('tmp_huge_plot.pkl','rb') as fid:
    fig = pickle.load(fid)
    
    
threshold_roc=0.034801
threshold_pr=0.481974
tpr=0.850000
fpr=0.196970

precision=0.209302
recall=0.270000

plt.gcf().get_axes()[0].plot([-2, 2], [threshold_roc, threshold_roc], color=cmap(0), alpha=0.9)
plt.gcf().get_axes()[0].plot([-2, 2], [threshold_pr, threshold_pr], color=cmap(0.99), alpha=0.9)

# ROC
plt.gcf().get_axes()[1].plot([0, 1], [tpr, tpr], color=cmap(0), alpha=0.9)
plt.gcf().get_axes()[1].plot([fpr, fpr], [0, 1], color=cmap(0), alpha=0.9)
plt.gcf().get_axes()[1].plot([0, 1], [recall, recall], color='black', alpha=0.5, linewidth=0.5)

# PR
plt.gcf().get_axes()[2].plot([0, 1], [precision, precision], color=cmap(0.99), alpha=0.9)
plt.gcf().get_axes()[2].plot([recall, recall], [0, 1], color=cmap(0.99), alpha=0.9)
plt.gcf().get_axes()[2].plot([0, 1], [tpr, tpr], color='black', alpha=0.5, linewidth=0.5)

plt.savefig('huge_example.png', bbox_inches='tight')
plt.show()
```
